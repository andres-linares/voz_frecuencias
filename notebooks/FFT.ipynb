```python
import pickle
import numpy as np
import matplotlib.pyplot as pyplot
import pandas as pd

FRECUENCIA = 16000
NOMBRE = 'andres/arriba'
```

```python
with open('grabaciones/{}.nvoice'.format(NOMBRE), 'rb') as handle:
    grabaciones = pickle.load(handle)
```

```python
transformadas = []
malos = []
if NOMBRE == 'vanessa/izquierda':
    malos = [0, 2, 15, 23]
elif NOMBRE == 'vanessa/derecha':
    malos = [0, 2, 6, 8, 10]
elif NOMBRE == 'vanessa/arriba':
    malos = [0, 2, 3, 8, 9, 10, 19, 23]
elif NOMBRE == 'vanessa/abajo':
    malos = [0, 13]
elif NOMBRE == 'vanessa/vanessa':
    malos = [0, 1, 3, 8, 9, 14, 16, 17, 18, 19, 25, 28]
elif NOMBRE == 'andres/izquierda':
    malos = [14, 28]
elif NOMBRE == 'andres/derecha':
    malos = [1, 25]
elif NOMBRE == 'andres/arriba':
    malos = [0, 3]
elif NOMBRE == 'andres/abajo':
    malos = [0, 1, 9, 21]
elif NOMBRE == 'andres/andres':
    malos = [0, 3, 13, 15, 19, 20, 21, 22]
for i in range(0 , len(grabaciones)):
    if i in malos:
        pass
    else:
        transformacion = np.abs(np.fft.fft(grabaciones[i]))
        transformadas.append(transformacion)
pyplot.plot(grabaciones[29])
pyplot.show()
```

```python
def obtener_energia(a, b, pos):
    energia = 0
    shift_mitad = len(transformacion) / 2
    escalador = shift_mitad / (FRECUENCIA / 2)
    extremo_izq = int(a * escalador)
    extremo_der = int(b * escalador)
    filtro = transformadas[pos][extremo_izq:extremo_der]
    energia = np.sum(filtro)
    return energia
```

```python
muestras_filtradas = []
for i in range(0, len(transformadas)):
    filtros = []
    filtros.append(obtener_energia(20, 100, i))
    filtros.append(obtener_energia(100, 200, i))
    filtros.append(obtener_energia(200, 300, i))
    filtros.append(obtener_energia(300, 400, i))
    filtros.append(obtener_energia(400, 500, i))
    filtros.append(obtener_energia(500, 600, i))
    filtros.append(obtener_energia(600, 700, i))
    filtros.append(obtener_energia(700, 800, i))
    filtros.append(obtener_energia(800, 900, i))
    filtros.append(obtener_energia(900, 1000, i))
    filtros.append(obtener_energia(1000, 1100, i))
    filtros.append(obtener_energia(1100, 1200, i))
    filtros.append(obtener_energia(1200, 1300, i))
    filtros.append(obtener_energia(1300, 1400, i))
    filtros.append(obtener_energia(1400, 1500, i))
    filtros.append(obtener_energia(1500, 1600, i))
    filtros.append(obtener_energia(1600, 1700, i))
    filtros.append(obtener_energia(1700, 1800, i))
    filtros.append(obtener_energia(1800, 1900, i))
    filtros.append(obtener_energia(1900, 2000, i))
    filtros.append(NOMBRE)
    muestras_filtradas.append(filtros)
muestra = pd.DataFrame(muestras_filtradas, columns=[
    '20-100 Hz',
    '100-200 Hz',
    '200-300 Hz',
    '300-400 Hz',
    '400-500 Hz',
    '500-600 Hz',
    '600-700 Hz',
    '700-800 Hz',
    '800-900 Hz',
    '900-1000 Hz',
    '1000-1100 Hz',
    '1100-1200 Hz',
    '1200-1300 Hz',
    '1300-1400 Hz',
    '1400-1500 Hz',
    '1500-1600 Hz',
    '1600-1700 Hz',
    '1700-1800 Hz',
    '1800-1900 Hz',
    '1900-2000 Hz',
    'Palabra'
])
muestra
```
